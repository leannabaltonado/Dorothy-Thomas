```python
AGE_BINS_5Y = [
    "0004","0509","1014","1519","2024","2529","3034","3539",
    "4044","4549","5054","5559","6064","6569","7074","7579","80UP"
]

AGE_LABEL = {
    "0004": "00-04", "0509": "05-09", "1014": "10-14", "1519": "15-19",
    "2024": "20-24", "2529": "25-29", "3034": "30-34", "3539": "35-39",
    "4044": "40-44", "4549": "45-49", "5054": "50-54", "5559": "55-59",
    "6064": "60-64", "6569": "65-69", "7074": "70-74", "7579": "75-79",
    "80UP": "80+",
}

```

```python
from typing import Literal

Sex = Literal["male", "female", "total"]
Metric = Literal["count", "pct"]  # count or percent-of-sex (5Y)

def wb_indicator_5y(age_code: str, sex: Literal["male","female"], metric: Metric) -> str:
    if age_code not in AGE_LABEL:
        raise ValueError(f"Unsupported age bin: {age_code}")

    sex_code = "MA" if sex == "male" else "FE"

    if metric == "count":
        return f"SP.POP.{age_code}.{sex_code}"
    if metric == "pct":
        return f"SP.POP.{age_code}.{sex_code}.5Y"

    raise ValueError("metric must be 'count' or 'pct'")
```

```python
import pandas as pd
import wbdata

def population_df(
    years: tuple[int, int],
    age_bins: list[str],
    area: str,
    sex: Sex = "total",
    metric: Metric = "count",
    keep_components: bool = False,
    use_pretty_labels: bool = True,
) -> pd.DataFrame:
    country_id = wbdata.get_countries(query=area)[0]["id"]

    # If total, fetch both sexes then sum (only makes sense for metric="count")
    sexes_to_fetch = ["male", "female"] if sex == "total" else [sex]

    indicators = {}
    for age_code in age_bins:
        for s in sexes_to_fetch:
            ind = wb_indicator_5y(age_code, s, metric)
            label = AGE_LABEL[age_code] if use_pretty_labels else age_code
            indicators[ind] = f"{s}_{label}"

    df = wbdata.get_dataframe(indicators, country=country_id, parse_dates=True)

    start, end = years
    df = df[(df.index.year >= start) & (df.index.year <= end)]

    if sex == "total":
        if metric != "count":
            raise ValueError("sex='total' with metric='pct' is not meaningful by simple addition. Use metric='count' or compute weighted %.")

        for age_code in age_bins:
            label = AGE_LABEL[age_code] if use_pretty_labels else age_code
            m = f"male_{label}"
            f = f"female_{label}"
            df[f"total_{label}"] = df[m] + df[f]

        if not keep_components:
            drop_cols = []
            for age_code in age_bins:
                label = AGE_LABEL[age_code] if use_pretty_labels else age_code
                drop_cols += [f"male_{label}", f"female_{label}"]
            df = df.drop(columns=drop_cols)

    return df
```

```python
import wbdata
import pandas as pd

def population_df(years, age_range, area) -> pd.DataFrame: 
    """
    years (tuple): (start_year, end_year), use (year, year) if just want singular year
    age_range (tuple): (low_age, high_age)
    area (str): e.g. "India"

    This function takes in year, age_range, and area [country/world] and outputs a DataFrame with relevant information
    """
    country_id = wbdata.get_countries(query=area)[0]["id"]

    low, high = age_range
    ages = range(low, high + 1)

    indicators = {}
    for age in ages:
        age2 = f"{age:02d}"
        indicators[f"SP.POP.AG{age2}.MA.IN"] = f"male_{age2}"
        indicators[f"SP.POP.AG{age2}.FE.IN"] = f"female_{age2}"

    df = wbdata.get_dataframe(indicators, country=country_id, parse_dates=True)

    start, end = years
    df = df[(df.index.year >= start) & (df.index.year <= end)]

    return df
```

```python
def population(years, sex, age_range, area):
    """
    years (tuple): (start_year, end_year) ,use (year, year) if just want singular year
    sex (str): "Male", "Female", or "All"
    age_range (tuple): (low_age, high_age)
    area (str): e.g. "India", "World"

    This function answers the question: 
    In [year] how many [people/males/females] aged [low] to [high] were living in [the world/region/country]?
    The output is a Series
    """
    df = population_df(years, age_range, area)
    df['male_total'] = df.filter(like='male_').sum(axis=1)
    df['female_total'] = df.filter(like='female_').sum(axis=1)
    df['population_total'] = df['female_total']+ df['male_total']
    
    #only the aggregated df
    agg = df[['male_total', 'female_total', 'population_total']]
    agg = agg.copy()
    agg.index = agg.index.year
    agg.index.name = "year"
    
    if sex == 'Female':
        return agg['female_total']
    elif sex == 'Male':
        return agg['male_total']
    elif sex == 'All':
        return agg['population_total']
    else:
        print("Invalid Input. Please input 'Female', 'Male', or 'All'.")
        
        

```

```python
#Unit test

#Spot check a few data to see if loaded df is correct


def spot_check_against_wb(df_loaded, country_id, indicators_dict, years_to_check):
    # pull fresh from WB for the same indicators
    df_fresh = wbdata.get_dataframe(indicators_dict, country=country_id, parse_dates=True)
    df_fresh = df_fresh[df_fresh.index.year.isin(years_to_check)]

    loaded = df_loaded[df_loaded.index.year.isin(years_to_check)][df_fresh.columns]
    
    # align rows
    loaded, fresh = loaded.align(df_fresh, join="inner", axis=0)

    diffs = (loaded - fresh).abs()
    max_abs = diffs.max().max()
    n_diff = (diffs > 0).sum().sum()  # exact mismatch count

    print("Max absolute difference:", max_abs)
    print("Number of nonzero diffs:", int(n_diff))

    # if floats ever appear, allow tolerance:
    # print("Within tol:", np.allclose(loaded.values, fresh.values, rtol=0, atol=1e-6))

    return diffs
```

```python
AGE_BINS_5Y = [
    "0004","0509","1014","1519","2024","2529","3034","3539",
    "4044","4549","5054","5559","6064","6569","7074","7579","80UP"
]

BIN_TO_RANGE = {
    "0004": (0,4),   "0509": (5,9),   "1014": (10,14), "1519": (15,19),
    "2024": (20,24), "2529": (25,29), "3034": (30,34), "3539": (35,39),
    "4044": (40,44), "4549": (45,49), "5054": (50,54), "5559": (55,59),
    "6064": (60,64), "6569": (65,69), "7074": (70,74), "7579": (75,79),
    "80UP": (80,200),
}
```

```python
def bins_for_age_range(start: int, end: int) -> list[str]:
    return [
        b for b, (lo, hi) in BIN_TO_RANGE.items()
        if not (hi < start or lo > end)
    ]
```

```python
def wb_indicator_5y(age_code: str, sex: str, metric: str) -> str:
    sex_code = "MA" if sex == "male" else "FE"
    return (
        f"SP.POP.{age_code}.{sex_code}"
        if metric == "count"
        else f"SP.POP.{age_code}.{sex_code}.5Y"
    )
```

```python
import time
import wbdata
import pandas as pd

def population_df(years, age_start, age_end, area, sex="total", metric="count"):
    age_bins = bins_for_age_range(age_start, age_end)
    country_id = wbdata.get_countries(query=area)[0]["id"]

    sexes = ["male", "female"] if sex == "total" else [sex]
    indicators = {
        wb_indicator_5y(b, s, metric): f"{s}_{b}"
        for b in age_bins
        for s in sexes
    }

    # retry once; JSONDecodeError is often transient
    for attempt in (1, 2):
        try:
            df = wbdata.get_dataframe(indicators, country=country_id, parse_dates=True)
            break
        except Exception:
            if attempt == 2:
                raise
            time.sleep(1.0)

    # If index is not datetime, convert it (often it's year strings)
    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index.astype(str), errors="coerce")

    y0, y1 = years
    df = df[(df.index.year >= y0) & (df.index.year <= y1)]

    if sex == "total":
        for b in age_bins:
            df[f"total_{b}"] = df[f"male_{b}"] + df[f"female_{b}"]
        df = df[[f"total_{b}" for b in age_bins]]

    return df
```

```python
import time
import pandas as pd
import wbdata

AGE_BINS_5Y = [
    "0004","0509","1014","1519","2024","2529","3034","3539",
    "4044","4549","5054","5559","6064","6569","7074","7579","80UP"
]

BIN_TO_RANGE = {
    "0004": (0,4),   "0509": (5,9),   "1014": (10,14), "1519": (15,19),
    "2024": (20,24), "2529": (25,29), "3034": (30,34), "3539": (35,39),
    "4044": (40,44), "4549": (45,49), "5054": (50,54), "5559": (55,59),
    "6064": (60,64), "6569": (65,69), "7074": (70,74), "7579": (75,79),
    "80UP": (80,200),
}

def bins_for_age_range(start_age: int, end_age: int) -> list[str]:
    ''' >>> bins_for_age_range(start_age = 0, end_age = 9)
        return: ['0004', '0509']
        '''
    return [b for b,(lo,hi) in BIN_TO_RANGE.items() if not (hi < start_age or lo > end_age)]
```

```python
def wb_indicator_5y(age_code: str, sex: str) -> str:
    ''' This helper constructs variable indicator.
    >>> wb_indicator_5y(age_code = '0004', sex = 'female')
    output: "SP.POP.0004.FE"
    '''
    
    sex_code = "FE" if sex == "female" else "MA"
    return f"SP.POP.{age_code}.{sex_code}"
```

```python
import re

def population_df(years: tuple[int,int], area: str, chunk: int = 4) -> pd.DataFrame:
    country_id = wbdata.get_countries(query=f"^{re.escape(area)}$")[0]["id"]

    def fetch(sub_bins):
        '''indicators = {
              "SP.POP.1014.FE": "female_1014",
              "SP.POP.1014.MA": "male_1014"
            }
            '''
        indicators = {}
        for b in sub_bins:
            indicators[wb_indicator_5y(b, "female")] = f"female_{b}"
            indicators[wb_indicator_5y(b, "male")]   = f"male_{b}"

        last_err = None
        for k in range(5):
            try:
                return wbdata.get_dataframe(indicators, country=country_id, parse_dates=True, skip_cache=True)
            except Exception as e:
                last_err = e
                time.sleep(0.5 * (2 ** k))
        raise last_err

    parts = []
    for i in range(0, len(AGE_BINS_5Y), chunk):
        parts.append(fetch(AGE_BINS_5Y[i:i+chunk]))

    df = pd.concat(parts, axis=1)

    if not isinstance(df.index, pd.DatetimeIndex):
        df.index = pd.to_datetime(df.index.astype(str), errors="coerce")

    y0, y1 = years
    df = df[(df.index.year >= y0) & (df.index.year <= y1)]

    # consistent column order: female bins then male bins
    df = df[[f"female_{b}" for b in AGE_BINS_5Y] + [f"male_{b}" for b in AGE_BINS_5Y]]

    df.index = df.index.year

    return df
```

```python
# Population Stats function


def population_stats(
    years: tuple[int, int],
    sex: str,
    age_range: tuple[int, int],
    area: str
) -> pd.DataFrame:
    
    df = population_df(years, area)  # full bins: female_0004..female_80UP + male_...

    start_age, end_age = age_range
    bins = bins_for_age_range(start_age, end_age)  # uses your BIN_TO_RANGE mapping

    if sex not in {"female", "male", "total"}:
        raise ValueError("sex must be 'female', 'male', or 'total'")

    if sex in {"female", "male"}:
        cols = [f"{sex}_{b}" for b in bins]
        out = df[cols].copy()
        out[f"{sex}_total"] = out.sum(axis=1)
        return out

    # sex == "total": derive as male+female
    female_cols = [f"female_{b}" for b in bins]
    male_cols   = [f"male_{b}" for b in bins]

    out = pd.DataFrame(index=df.index)
    out["female_total"] = df[female_cols].sum(axis=1)
    out["male_total"]   = df[male_cols].sum(axis=1)
    out["total"]        = out["female_total"] + out["male_total"]
    return out
```
